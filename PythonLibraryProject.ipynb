# **Descriptive analysis for those whose Job Role is Manager**

```python
manager_role = data[data['JobRole']=='Manager']
manager_role.head(5)
```

```python
manager_role.describe()
```

```python
import plotly.express as px
fig = px.box(manager_role, y='JobSatisfaction')
fig.show()
```

```python
fig = px.scatter(manager_role, x="WorkLifeBalance",y='JobSatisfaction')
fig.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,7))
sns.barplot(x='EducationField',y='JobSatisfaction',hue='JobRole',data=manager_role)
plt.xticks(rotation=90)
plt.title("Educationfiled v/s JobSatisfaction")
plt.show()
```

```python
new_df = manager_role.groupby(['Age'])
new_df.first()
```
